```python
import pandas as pd
ds=pd.read_csv("diabetes.csv",header=None,skipinitialspace=True,skiprows=1)
X=ds.iloc[:,:-1]
y=ds.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
X_train
import seaborn as sns
sns.heatmap(X.isnull())
```

```python
from keras.models import Sequential
mind=Sequential()
from keras.layers import Dense
mind.add(Dense(units=4,activation="relu",kernel_initializer='zeros',bias_initializer='zeros',input_dim=8))
mind.add(Dense(units=4,activation="relu",kernel_initializer='zeros',bias_initializer='zeros'))
mind.add(Dense(units=1,activation="sigmoid",kernel_initializer='zeros',bias_initializer='zeros'))
```

```python
mind.summary()
```

```python
from keras.optimizers import Adam
mind.compile(loss="binary_crossentropy",optimizer=Adam(learning_rate=0.001))
X_train.shape
```

```python
#import time
mind.fit(X_train,y_train,epochs=100)
#time.sleep(0.1)
```
